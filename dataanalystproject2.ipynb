```python
# write simple code for barchat
import itertools
import matplotlib.pyplot as plt
# Get the number of sales for each state
sales_by_state = df.groupby('State')['Amount'].sum()

colormap = plt.cm.get_cmap('tab20', len(sales_by_state))  # 'tab20' can handle up to 20 different colors

# Get unique colors for each bar
colors = [colormap(i) for i in range(len(sales_by_state))]
# Create a bar chart of the sales by state
plt.bar(sales_by_state.index, sales_by_state.values, color=colors)
plt.xticks(rotation=60)
for index, value in enumerate(sales_by_state.values):
    plt.text(index, value, str(value), ha='center', va='bottom', rotation=90)  # Rotate the label by 45 degrees

# Adjust the layout to ensure labels fit
plt.tight_layout()
plt.ylim(0,30000000)
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.title('Total Sales by State')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter data for males
df_male = df[df['Gender'] == 'M']
x_male = df_male["Age"]
y_male = df_male["Amount"]

# Filter data for females
df_female = df[df['Gender'] == 'F']
x_female = df_female["Age"]
y_female = df_female["Amount"]

# Plot male data
plt.plot(x_male, y_male, "r*", label='Male expenditure')

# Plot female data
plt.plot(x_female, y_female, "b*", label='Female expenditure')

# Add labels and title
plt.xlabel("Age")
plt.ylabel("Amount")
plt.title("Age vs Expenditure by Gender")

# Add legend
plt.legend()

# Show the plot
plt.show()
```
